```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("student_performance_prediction.csv")
```

```python
df.head()
```

```python
df = df.drop(columns=["Student ID", 
                      "Attendance Rate", 
                      "Participation in Extracurricular Activities", 
                      "Parent Education Level"])
df = df.dropna()
```

```python
df.head()
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
```

```python
df['Passed'] = df['Passed'].replace({'Yes': 1, 'No': 0})
plt.scatter(df['Study Hours per Week'],df['Previous Grades'],c=df['Passed'])
```

```python
x = df.iloc[:,0:2]
```

```python
y = df.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.preprocessing import StandardScaler
scalar =StandardScaler()
x_train = scalar.fit_transform(x_train)
```

```python
x_train
```

```python
x_test = scalar.transform(x_test)
x_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
cif = LogisticRegression()
```

```python
cif.fit(x_train , y_train)
```

```python
 y_pred =cif.predict(x_test)
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x_train,y_train.values,clf=cif,legend=2)
```
